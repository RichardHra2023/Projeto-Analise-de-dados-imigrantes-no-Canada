###Realizando alguns tratamentos

```python
# Setando o nome do país como indice do DataFrame

df.set_index('País', inplace=True)
```

```python
df
```

```python
# criando uma lista chamada anos que contém strings representando os anos de 1980 a 2013

anos = list(map(str, range(1980, 2014)))
```

```python
anos
```

## Extraindo a série de dados para o Brasil

```python
brasil = df.loc['Brasil', anos]
```

```python
#Dados com a quantidade de imigrantes do Brasil por ano:
brasil
```

```python
# Criando um DataFrame do Brasil

brasil_dict = {'Ano': brasil.index.tolist(), 'Numero de Imigrantes': brasil.values.tolist()}
dados_brasil = pd.DataFrame(brasil_dict)
```

```python
dados_brasil = pd.DataFrame(brasil_dict)
```

```python
dados_brasil
#Dados feitos com imigrantes feitos por tabelas sobre a quantidade de imigrantes.
```

```python
# Gráfico da representação da quantidade de imigrantes brasileiros no canadá

dados_brasil.plot.line ('Ano', 'Numero de Imigrantes', title = 'Quantidade de imigrantes brasileiros no canadá')
```
